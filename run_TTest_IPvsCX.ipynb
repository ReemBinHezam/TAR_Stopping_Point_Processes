```python
# Importing library
import scipy.stats as stats
import pandas as pd
import numpy as np
```

```python
#set path
PARENT_DIR = '/content/drive/MyDrive/ReemSharedPhD/SharedCode/TOIS_paper_code/'
```

# Load data

```python


# prepare data for TTest
df_all = pd.read_csv(PARENT_DIR+'TTest/results/df_all_rates_forTTest.csv', index_col=0)
df_all['Model'] = df_all.index
df_all['process_type'] = df_all['Model'].apply(lambda x: 'IP' if any(i in x for i in ['IP']) else 'CX')
df_all = df_all.drop(['Model'], axis=1)
df_all
```

```python
df_all_IP = df_all.loc[df_all['process_type'] == 'IP']
df_all_CX = df_all.loc[df_all['process_type'] == 'CX']

df_all_IP.index.name='Model'
df_all_IP = df_all_IP.reset_index()

df_all_CX.index.name='Model'
df_all_CX = df_all_CX.reset_index()
```

## Paired T-Test (for AutoTAR ranking on all datasets)

From results, the difference between the Inhomogeneous Poisson and Cox
processe found to be statistically significant (< 0.05) for all measures expet cost.
In average, the Inhomogeneous Poisson process achieved higher recall and reliability, with less relative error than the Cox process. However, although the Inhomogeneous Poisson process had higher cost than the Cox process, the difference between them were not statistically significiant. Thus Inhomogeneous Poisson process would be selected among the the Cox process.

### T-Test for all target recalls

```python
eval_measures = ['recall', 'cost', 'reliability', 'loss_er','rel_err']
for eval_measure in eval_measures:
    statistic, pvalue = stats.ttest_rel(df_all_IP[eval_measure], df_all_CX[eval_measure])
    print(f'{eval_measure: <15}: statistic=\t{statistic: <20}\t pvalue= {pvalue}')
```

```python
print('IP recall is higher than CX recall with an average: ', np.mean(df_all_IP['recall']- df_all_CX['recall']))
print('IP cost is higher than CX cost with an average: ', np.mean(df_all_IP['cost']- df_all_CX['cost']))
print('IP reliability is higher than CX reliability with an average: ', np.mean(df_all_IP['reliability']- df_all_CX['reliability']))
print('IP loss_er is lower than CX loss_er with an average: ', np.mean(df_all_IP['loss_er']- df_all_CX['loss_er']))
print('IP rel_err is lower than CX rel_err with an average: ', np.mean(df_all_IP['rel_err']- df_all_CX['rel_err']))
```

### T-Test for each target recall

```python
des_recalls = [1.0, 0.9, 0.8]
eval_measures = ['recall', 'cost', 'reliability', 'loss_er','rel_err']
for des_recall in des_recalls:
  for eval_measure in eval_measures:
    statistic, pvalue = stats.ttest_rel(df_all_IP.loc[df_all_IP['des_recall'] == des_recall][eval_measure], df_all_CX.loc[df_all_CX['des_recall'] == des_recall][eval_measure])
    print(f'{des_recall}_{eval_measure: <15}: statistic=\t{statistic: <20}\t pvalue= {pvalue}')
  print('-----------------------------------------------------------------------------------')
    #print(des_recall, eval_measure, stats.ttest_rel(df_all_IP.loc[df_all_IP['des_recall'] == des_recall][eval_measure], df_all_CX.loc[df_all_CX['des_recall'] == des_recall][eval_measure]))
```
